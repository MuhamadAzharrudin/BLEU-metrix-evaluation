```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statistics import mean, median, stdev

# --- Load Data ---
df = pd.read_csv("hasil_parafrase.csv")

# --- Hitung Metrik BLEU ---
bleu_scores = df['bleu_score'].tolist()

avg_bleu = mean(bleu_scores)
median_bleu = median(bleu_scores)
std_bleu = stdev(bleu_scores)

print("ðŸ“Š Evaluasi BLEU:")
print(f"ðŸ”¹ Rata-rata BLEU: {avg_bleu:.4f}")
print(f"ðŸ”¹ Median BLEU: {median_bleu:.4f}")
print(f"ðŸ”¹ Standard Deviasi: {std_bleu:.4f}")
print(f"ðŸ”¹ Skor BLEU Tertinggi: {max(bleu_scores):.4f}")
print(f"ðŸ”¹ Skor BLEU Terendah: {min(bleu_scores):.4f}")

# --- Plot Distribusi BLEU ---
plt.figure(figsize=(10, 5))
sns.histplot(bleu_scores, bins=10, kde=True, color='skyblue')
plt.title("Distribusi Skor BLEU")
plt.xlabel("Skor BLEU")
plt.ylabel("Jumlah Sample")
plt.grid(True)
plt.tight_layout()
plt.show()
```
